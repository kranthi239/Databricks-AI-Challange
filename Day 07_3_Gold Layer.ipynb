# Gold Layer (Dependent on Silver)

```python
dbutils.widgets.text("event_date", "")
dbutils.widgets.text("silver_path", "/Volumes/workspace/ecommerce/ecommerce_data/Delta/silver_nov/")
dbutils.widgets.text("gold_path", "/Volumes/workspace/ecommerce/ecommerce_data/Delta/gold_nov/")
```

# Read Widget Values

```python
event_date = dbutils.widgets.get("event_date")
silver_path = dbutils.widgets.get("silver_path")
gold_path = dbutils.widgets.get("gold_path")
```

# Read Silver layer

```python
df_silver = spark.read.format("delta").load(silver_path)
```

# Data Aggregation 

```python
df_gold = (
    df_silver
    .groupBy("category_code")
    .count()
)
```

# Write Gold layer

```python
df_gold.write.format("delta")\
    .mode("overwrite")\
    .save(gold_path)
```

```python
print(f"Gold aggregation completed for date: {event_date}")
print(f"Gold records: {df_gold.count()}")
```
