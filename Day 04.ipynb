```python
from pyspark.sql.functions import *
```

# Creating Dataframe from Sample Oct and Nov Data

```python
df_nov = spark.read.csv("/Volumes/workspace/ecommerce/ecommerce_data/2019-Oct.csv", header = True, inferSchema = True)
df_oct = spark.read.csv("/Volumes/workspace/ecommerce/ecommerce_data/2019-Nov.csv", header = True, inferSchema = True)
```

# Convert CSV DataFrames to Delta Format

```python
df_oct.write.format("delta")\
    .mode("overwrite")\
    .save("/Volumes/workspace/ecommerce/ecommerce_data/Delta/event_oct/")

df_nov.write.format("delta")\
    .mode("overwrite")\
    .save("/Volumes/workspace/ecommerce/ecommerce_data/Delta/event_nov/")
```

# Reading Delta tables

```python
%sql
SELECT * FROM delta.`/Volumes/workspace/ecommerce/ecommerce_data/Delta/event_oct/`
```

# Create Delta Tables (SQL + PySpark)

```python
%sql
CREATE TABLE workspace.ecommerce.events_oct
USING DELTA;
```

```python
df_oct.write.mode("overwrite").option("mergeSchema", "true").saveAsTable("workspace.ecommerce.events_oct")
```
