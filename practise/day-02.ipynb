```python

# Load data
events = spark.read.csv("/Volumes/workspace/ecommerce/ecommerce_data/2019-Oct.csv", header=True, inferSchema=True)

# Selecting specific columns
print("\n Selecting specific columns (NOTE: No data will be displayed as we are only using transformations and not actions)")
events.select("event_type", "product_id", "price")

#displaying specific columns
print("\n Selecting specific columns and showcasing lazy evaluation concept.(NOTE: Data will be displayed as we are using actions too) ")
events.select("event_type", "product_id", "price").show(10)

#Filtering
print("\nPrinting the count of the records whose price is greater than 100")
x = events.filter("price > 100").count()
print("Count = ",x)

#groupby
print("\nGrouping the data by event type")
events.groupBy("event_type").count().show()

#order by and limit
print("\nPrinting the top 5 brands based on the count of the records")
top_brands = events.groupBy("brand").count().orderBy("count", ascending=False).limit(5)
top_brands.show()
```
