# Load November 2019 Data

```python
nov_events = spark.read.csv("/Volumes/workspace/ecommerce/ecommerce_data/2019-Nov.csv", header=True, inferSchema=True)
nov_events.write.mode("append").saveAsTable("workspace.ecommerce.events")
```

# Load October 2019 Data

```python
oct_events = spark.read.csv("/Volumes/workspace/ecommerce/ecommerce_data/2019-Oct.csv", header=True, inferSchema=True)
oct_events.write.mode("append").saveAsTable("workspace.ecommerce.events")
```

# Display Dataset Statistics and Schema

```python
print(f"October 2019 - Total Events: {oct_events.count():,}")
print("\n" + "="*60)
print("SCHEMA:")
print("="*60)
oct_events.printSchema()
```
