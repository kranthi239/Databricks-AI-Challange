```python
from pyspark.sql.functions import *
```

# Loading Nov data into DataFrame

```python
df_nov = spark.read.csv("/Volumes/workspace/ecommerce/ecommerce_data/2019-Oct.csv", header=True, inferSchema=True)
df_nov.display()
```

# Filter rows with Event Type is "view"

```python
df_event = df_nov.filter(df_nov.event_type == 'view')
df_event.display()
```

# Select Relevent columns when Event type is "View"

```python
df_selected = df_event.select('event_time', 'user_id', 'product_id')
df_selected.display()
```

# Calculate Top 5 Product Sales for Event Type is "View"

```python
df_topSales = df_event.filter(df_event.event_type == "view"
        ).groupBy("product_id").agg(
            sum("price").alias("Product_Sales")
        ).orderBy("product_Sales", ascending=False).limit(5)

df_topSales.display()
```

# Saving Top Sales Data in Delta Table

```python
df_topSales.write.format("csv").mode("overwrite").save("/Volumes/workspace/ecommerce/ecommerce_data/topSales")
```
