# Add Widgets for parameters

```python
dbutils.widgets.text("event_date","")
dbutils.widgets.text("source_path","/Volumes/workspace/ecommerce/ecommerce_data/2019-Nov.csv")
dbutils.widgets.text("bronze_path1","/Volumes/workspace/ecommerce/ecommerce_data/Delta/bronze_nov/")

```

# Reading Widget Value

```python
event_date = dbutils.widgets.get("event_date")
source_path = dbutils.widgets.get("source_path")
bronze_path = dbutils.widgets.get("bronze_path1")
print(f"Bronze Data loaded at : {bronze_path}")
```

# Read Raw Data

```python
df_raw = spark.read.format("csv")\
          .options(header="true", inferSchema="true")\
          .load(f"{source_path}/{event_date}")
df_raw.display()
```

# Writing Raw data into Bronze layer in Delta format 

```python
df_raw.write.format("delta")\
    .mode("append")\
    .save(bronze_path)
```

# Logging the values

```python
print(f"Bronze load completed for date : {event_date}")
print(f"Bronze Data loaded at : {bronze_path}")
```
