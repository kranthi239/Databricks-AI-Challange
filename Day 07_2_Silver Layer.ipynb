# Silver Layer (Dependent on Bronze)

# Adding Widgets for Silver layer

```python
dbutils.widgets.text("event_date","")
dbutils.widgets.text("bronze_path","/Volumes/workspace/ecommerce/ecommerce_data/Delta/bronze_nov/")
dbutils.widgets.text("silver_path","/Volumes/workspace/ecommerce/ecommerce_data/Delta/silver_nov/")
```

# Read Widget Values

```python
event_date = dbutils.widgets.get("event_date")
bronze_path = dbutils.widgets.get("bronze_path")
silver_path = dbutils.widgets.get("silver_path")
```

# Read Bronze Delta

```python
df_bronze = spark.read.format("delta").load(bronze_path)
```

```python
df_silver = (df_bronze
            .dropDuplicates()
            .filter("event_type is Not Null"))
```

%md
# Writing Bronze data into Silver layer in Delta format 

```python
df_silver.write.format("delta")\
        .mode("append")\
        .save(silver_path)
```

```python
print(f"Silver transformation completed for date: {event_date}")
print(f"Records in silver: {df_silver.count()}")
```
